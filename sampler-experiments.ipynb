```python
from edict.edict import coupled_stablediffusion
from edict.utils import load_idx
```

```python
with torch.no_grad():
    element_idx = 0
    element = load_idx(data, element_idx, model.device)
    closest_element_idx = 330
    closest_element = load_idx(data, closest_element_idx, model.device)
    conditioner_input_keys = [e.input_key for e in model.conditioner.embedders]
    ucg_keys = conditioner_input_keys
    x = element[model.input_key]
    x = x.to(model.device)
    N = x.shape[0]
    x = x.flatten(0, 1)
    x_closest = closest_element[model.input_key]
    x_closest = x_closest.to(model.device).flatten(0, 1)
    c, uc = model.conditioner.get_unconditional_conditioning(
        element,
        force_uc_zero_embeddings=ucg_keys
        if len(model.conditioner.embedders) > 0
        else [],
        )
    z_first_stage = model.encode_first_stage(x)
    z_first_stage_closest = model.encode_first_stage(x_closest)
```

```python
latents = coupled_stablediffusion(
    model,
    init_image_latents=z_first_stage,
    conditioning_conditional=c,
    conditioning_unconditional=uc,
    reverse=True,
    device="cuda:3",
    sampling_kwargs={
        "image_only_indicator": element["image_only_indicator"],
    }
)
latentslatents = coupled_stablediffusion(
    model,
    init_image_latents=z_first_stage,
    fixed_starting_latent=latents,
    conditioning_conditional=c,
    conditioning_unconditional=uc,
    reverse=False,
    device="cuda:3",
    sampling_kwargs={
        "image_only_indicator": element["image_only_indicator"],
    }
)
```

```python
latents_closest = coupled_stablediffusion(
    model,
    init_image_latents=z_first_stage_closest,
    conditioning_conditional=c,
    conditioning_unconditional=uc,
    reverse=True,
    device="cuda:3",
    sampling_kwargs={
        "image_only_indicator": element["image_only_indicator"],
    }
)
latentslatents_closest = coupled_stablediffusion(
    model,
    init_image_latents=z_first_stage_closest,
    fixed_starting_latent=latents_closest,
    conditioning_conditional=c,
    conditioning_unconditional=uc,
    reverse=False,
    device="cuda:3",
    sampling_kwargs={
        "image_only_indicator": element["image_only_indicator"],
    }
)
```

```python
latents_baseline = coupled_stablediffusion(
    model,
    init_image_latents=z_first_stage,
    conditioning_conditional=c,
    conditioning_unconditional=uc,
    reverse=True,
    run_baseline=True,
    device="cuda:3",
    sampling_kwargs={
        "image_only_indicator": element["image_only_indicator"],
    }
)
latentslatents_baseline = coupled_stablediffusion(
    model,
    init_image_latents=z_first_stage,
    fixed_starting_latent=latents_baseline[0],
    conditioning_conditional=c,
    conditioning_unconditional=uc,
    reverse=False,
    run_baseline=True,
    device="cuda:3",
    sampling_kwargs={
        "image_only_indicator": element["image_only_indicator"],
    }
)
```

```python
latents_closest_baseline = coupled_stablediffusion(
    model,
    init_image_latents=z_first_stage_closest,
    conditioning_conditional=c,
    conditioning_unconditional=uc,
    reverse=True,
    run_baseline=True,
    device="cuda:3",
    sampling_kwargs={
        "image_only_indicator": element["image_only_indicator"],
    }
)
latentslatents_closest_baseline = coupled_stablediffusion(
    model,
    init_image_latents=z_first_stage_closest,
    fixed_starting_latent=latents_closest_baseline[0],
    conditioning_conditional=c,
    conditioning_unconditional=uc,
    reverse=False,
    run_baseline=True,
    device="cuda:3",
    sampling_kwargs={
        "image_only_indicator": element["image_only_indicator"],
    }
)
```

```python
decode_z = model.decode_first_stage(z_first_stage)
decode_z_closest = model.decode_first_stage(z_first_stage_closest)
decode0 = model.decode_first_stage(latentslatents[0])
decode1 = model.decode_first_stage(latentslatents[1])
decode0_noised = model.decode_first_stage(latents[0])
decode1_noised = model.decode_first_stage(latents[1])
decode0_closest = model.decode_first_stage(latentslatents_closest[0])
decode1_closest = model.decode_first_stage(latentslatents_closest[1])
decode0_closest_noised = model.decode_first_stage(latents_closest[0])
decode1_closest_noised = model.decode_first_stage(latents_closest[1])
decode_baseline = model.decode_first_stage(latentslatents_baseline[0])
decode_closest_baseline = model.decode_first_stage(latentslatents_closest_baseline[0])
```
